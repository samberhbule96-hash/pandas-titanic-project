Part 2: Titanic Dataset Analysis

```python
file_path = '/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv'
df = pd.read_csv(file_path)

print(df.head())
df.info()
print(df.describe())
```

Step 2: Data Cleaning

This step fixes the missing values and removes columns that are too messy to use.

```python
# 1. Fill Age with median (the middle value)
df['Age'] = df['Age'].fillna(df['Age'].median())

# 2. Fill Embarked with the mode (the most common port)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 3. Drop Cabin (too many missing values)
df.drop(columns=['Cabin'], inplace=True)

# 4. Remove duplicate rows
df.drop_duplicates(inplace=True)

print("Cleaning Complete. Remaining missing values:")
print(df.isnull().sum())
```

Step 3: Data Analysis

Using groupby() to find patterns in survival.

```python
# Survival rate by Gender
print("Survival by Gender:\n", df.groupby('Sex')['Survived'].mean())

# Survival rate by Class
print("\nSurvival by Class:\n", df.groupby('Pclass')['Survived'].mean())

# Average age per Class
print("\nAvg Age per Class:\n", df.groupby('Pclass')['Age'].mean())

# --- FIXED SECTION ---
# Survival rate by Age Group (added observed=False to remove the warning)
bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

print("\nSurvival by Age Group:")
print(df.groupby('AgeGroup', observed=False)['Survived'].mean())
```

Step 4: Filtering

Extracting specific "at-risk" or "prioritized" groups.

```python
# Female passengers who survived
females_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]

# Children who survived
children_survived = df[(df['Age'] < 18) & (df['Survived'] == 1)]

# 1st class females (High survival probability)
top_tier_survival = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]

print(f"Number of Female Survivors: {len(females_survived)}")
print(f"Number of Child Survivors: {len(children_survived)}")
```
